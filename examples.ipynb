```python
from pprint import pprint
import numpy as np
from numpy.linalg import norm

um = np.array(
    [
        [5, 4, 2, 0, 0, 3],
        [5, 4, 4, 2, 0, 2],
        [4, 0, 3, 5, 2, 0],
        [4, 1, 3, 0, 0, 0],
        [4, 1, 3, 0, 0, 0],
    ]
)


def cosine_sim(v1, v2):
    return np.dot(v1, v2) / (norm(v1) * norm(v2))


sm = np.ndarray(shape=(um.shape[0], um.shape[0]))

for i in range(len(um)):
    for j in range(len(um)):
        sm[i][j] = cosine_sim(um[i], um[j])

print(sm.round(decimals=2))
```
